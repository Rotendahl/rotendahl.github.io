```julia
using Distributions
using Plots
using Random
```

```julia
using Pkg
Pkg.add("IJulia")
Pkg.add("Revise")
```

## Simple plot

```julia
plotly()
a, b = 2, 1
xs = range(-50, 50)
f(x) = a * x + b
noise = rand(Normal(0, 10), length(xs))
ys = f.(xs) + noise
Plots.scatter(xs, ys, label="data")
```

## Data Generator

```julia
function GenerateData(θ, n=100, σ=5)
  points = rand(Uniform(-50, 50), (n, length(θ)))
  points[:, length(θ)] .= 1
  ys = points * θ + rand(Normal(0, σ), n)
  points[:, length(θ)] = ys
  points
end

xs = GenerateData([2, 1, 3])
Plots.scatter(xs[:, 1], xs[:, 2], xs[:, 3], label="data")
```
